```python
import os
from urllib.request import urlretrieve
```

```python
# Dowload income data
data_url = "https://www.abs.gov.au/statistics/labour/jobs/jobs-australia/2020-21-financial-year/Table%201%20-%20Jobs%20and%20employment%20income%20by%20sex%2C%20age%2C%20business%20characteristics%20and%20geography%2C%202016-17%20to%202020-21.xlsx"
urlretrieve(data_url, "../data/landing/income.xlsx")
```

```python
import pandas as pd
df = pd.read_excel('../data/landing/income.xlsx', sheet_name='Table 1.5')

# Step 2: Rename the columns for better clarity
df.columns = ['SA2', 'SA2_Name', 'NUMBER_OF_JOBS_MALES_2016-17', 'NUMBER_OF_JOBS_MALES_2017-18', 
              'NUMBER_OF_JOBS_MALES_2018-19', 'NUMBER_OF_JOBS_MALES_2019-20', 'NUMBER_OF_JOBS_MALES_2020-21', 
              'NUMBER_OF_JOBS_FEMALES_2016-17', 'NUMBER_OF_JOBS_FEMALES_2017-18', 
              'NUMBER_OF_JOBS_FEMALES_2018-19', 'NUMBER_OF_JOBS_FEMALES_2019-20', 'NUMBER_OF_JOBS_FEMALES_2020-21',
              'NUMBER_OF_JOBS_PERSONS_2016-17', 'NUMBER_OF_JOBS_PERSONS_2017-18', 
              'NUMBER_OF_JOBS_PERSONS_2018-19', 'NUMBER_OF_JOBS_PERSONS_2019-20', 'NUMBER_OF_JOBS_PERSONS_2020-21',
              'MEDIAN_INCOME_MALES_2016-17', 'MEDIAN_INCOME_MALES_2017-18', 'MEDIAN_INCOME_MALES_2018-19', 
              'MEDIAN_INCOME_MALES_2019-20', 'MEDIAN_INCOME_MALES_2020-21', 'MEDIAN_INCOME_FEMALES_2016-17', 
              'MEDIAN_INCOME_FEMALES_2017-18', 'MEDIAN_INCOME_FEMALES_2018-19', 'MEDIAN_INCOME_FEMALES_2019-20', 
              'MEDIAN_INCOME_FEMALES_2020-21', 'MEDIAN_INCOME_PERSONS_2016-17', 'MEDIAN_INCOME_PERSONS_2017-18', 
              'MEDIAN_INCOME_PERSONS_2018-19', 'MEDIAN_INCOME_PERSONS_2019-20', 'MEDIAN_INCOME_PERSONS_2020-21']
```

```python
df.head()
```

```python
# Find the starting row for 'Victoria' and ending row for 'Queensland'
start_row = df[df['SA2'] == 'Victoria'].index[0]
end_row = df[df['SA2'] == 'Queensland'].index[0]

# Extract data of 'Victoria'
df = df.iloc[start_row + 1:end_row]
```

```python
df.head()
```

```python
# Drop unnecessary columns
df = df[['SA2_Name', 'NUMBER_OF_JOBS_PERSONS_2016-17', 'NUMBER_OF_JOBS_PERSONS_2017-18', 
              'NUMBER_OF_JOBS_PERSONS_2018-19', 'NUMBER_OF_JOBS_PERSONS_2019-20', 'NUMBER_OF_JOBS_PERSONS_2020-21', 
              'MEDIAN_INCOME_PERSONS_2016-17', 'MEDIAN_INCOME_PERSONS_2017-18', 'MEDIAN_INCOME_PERSONS_2018-19', 
              'MEDIAN_INCOME_PERSONS_2019-20', 'MEDIAN_INCOME_PERSONS_2020-21']]
```

```python
# Multiply Number of Jobs by 1000
df[['NUMBER_OF_JOBS_PERSONS_2016-17', 'NUMBER_OF_JOBS_PERSONS_2017-18', 
              'NUMBER_OF_JOBS_PERSONS_2018-19', 'NUMBER_OF_JOBS_PERSONS_2019-20', 'NUMBER_OF_JOBS_PERSONS_2020-21']] = df[['NUMBER_OF_JOBS_PERSONS_2016-17', 'NUMBER_OF_JOBS_PERSONS_2017-18', 
              'NUMBER_OF_JOBS_PERSONS_2018-19', 'NUMBER_OF_JOBS_PERSONS_2019-20', 'NUMBER_OF_JOBS_PERSONS_2020-21']] * 1000
```

```python
# Clean suburb names
pattern = r'^\s+|\s+$'
df['SA2_Name'] = df['SA2_Name'].str.replace(pattern, '', regex=True)
```

```python
# Assume 'np' values mean 0
df = df.replace(r'(np)+', 0, regex=True)
```

```python
df.to_csv('../data/curated/suburbs_income.csv', index=False)
```
